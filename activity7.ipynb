```python
 import pandas as pd

# Load the football dataset into a dataframe called "foot_ball"
foot_ball = pd.read_csv("C:/Users/CC/Downloads/dataset - 2020-09-24 (2).csv")

 
# Display the last 7 rows of the dataframe
print(foot_ball.tail(7))

# Display the last 7 rows of the dataframe
print(foot_ball.tail(7))

# Access and display the data for the tenth player in the dataset
print(foot_ball.iloc[9])

 #Access and display the "Goals" and "Appearances" for players with index positions 100 to 110
print(foot_ball.loc[100:110, ['Goals', 'Appearances']])

# Add a new column named "Goals per Appearance" and display the first 5 rows of the updated dataset
foot_ball['Goals per Appearance'] = foot_ball['Goals'] / foot_ball['Appearances']
print(foot_ball.head())

# Select and display a subset of the dataframe containing only the players from the "Arsenal" club
arsenal_players = foot_ball[foot_ball['Club'] == 'Arsenal']
print(arsenal_players)

# Perform a filtering operation to display players who have scored more than 5 goals
high_scorers = foot_ball[foot_ball['Goals'] > 5]
print(high_scorers)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting aesthetic style
sns.set_style("darkgrid")

# Scaling plots for different contexts
sns.set_context("notebook")




# Boxplot for goals distribution among clubs
sns.boxplot(x='Club', y='Goals', data=foot_ball)
plt.xticks(rotation=90)  # Rotating x-axis labels for better readability
plt.show()

# Violinplot for distribution of goals among different positions
sns.violinplot(x='Position', y='Goals', data=foot_ball)
plt.show()

 
```
